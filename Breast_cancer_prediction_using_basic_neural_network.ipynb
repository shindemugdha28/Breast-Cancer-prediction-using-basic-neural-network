**Sepreating features and targets**

```python
#Storing all the features in X data frame except coloumn name label. For Coloumn we use axis = 1 and for rows we use axis = 0
X = data_frame.drop(columns='label', axis=1)
#Stroring coloumn name lable into a Y dataframe
Y = data_frame['label']
# Using features from x dataframe we are going to predict values from Y dataframe
```

```python
print(X)
```

**Splitting the dataset into training and testing**

```python
# We are spliiting our test data in the ration of 80 and 20%, test size is 0.2 that mens we want 20% of data for testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# standardization makes data simpler to process, analyze, and store data in a database.When the feature distribution is consistent, it is helpful
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building Basic Neural Network**

```python
# importing tensorflow and Keras
#Tensor flow is the deep learning lib which is developed by google
import tensorflow as tf
#here we are setting random seed to 3 because to hget the same accuracy everytime we run the code you can set the seed value to any number
tf.random.set_seed(3)
# keras is not stand alone library , keras used tensar flow as its backend so that we can easily build a NN
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
# Created a varaiable name called model to store the neural network.
# Keras has prebuild function available called sequential with the help of that we can create a stack of layer sequentially
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)), #Input layer, with the help of Flattening  we will convert our input values in to a single dimension array before feeding the model, Input shape 30 is there because we have 30 features/ coloumns
                          keras.layers.Dense(20, activation='relu'), # Hidden Layer , We are taking 20 neurons, we can give any any number  and actuvation function
                          keras.layers.Dense(2, activation='sigmoid') # Output layer, we are menteioning two neurons here because our problems contains only 2 claases 0(Malignanat) and 1(benign), 1 neuron for each class. If the output is 1 then only one neuron which was dedicted to that value would be active other neuron wont be. If you would have got three classes then the neuron would have changed to 3
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Here we have labels like 1 for benign and 0 for malignant in this case we will use a sparse catogorical encoding.
If we have labels in one not enocing format like [1 0 0] for malignant and [0 1 0] for benign then we will use catogrical entropy. Like these there are more loss functions are available

```python
# training the neural Network
# epochs are how many time my model needs to go through the training data
# validation split is 0.1 that means 1 percent of data from the tarining data set
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

**Visualizing accuracy and loss**

```python
# For accuracy of validation data set and tarining dataset against each epoch
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
# Plotting for loss fucntion of valodation data and training data against each epoch
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```
